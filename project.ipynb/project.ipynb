# Cleaning Injury Severity
In this section we will categorize the injury severity into two which are 'fatal and 'non_fatal' to simplify the column for more clarity and easy understsnding on the level of severity of injuries.

```python

# Step 1: Standardize capitalization and strip spaces
df['Injury.Severity'] = df['Injury.Severity'].astype(str).str.strip().str.capitalize()

# Step 2: Create Grouped_injury column
def clean_injury(x):
    if 'Fatal' in x:          # catches Fatal, Fatal(1), Fatal(2), etc.
        return 'Fatal'
    elif x.lower() == 'nan':   # handle NaN values converted to string
        return 'Unknown'
    else:
        return 'Non-Fatal'

df['Grouped_injury'] = df['Injury.Severity'].apply(clean_injury)

# Step 3: Check counts
print("Grouped_injury counts:")
print(df['Grouped_injury'].value_counts())
```

From the above step we can conclude that the reported number of Non fatal injuries(70063) are more as compared to reported fatal injuries(17826) in the aircraft accidents.

# Group Engine Types
In this section we will classify the engine types into categories based on their unique shared properties to allow meaningful comparisons across the aircraft categories.We will use the most common engine types that appear in our dataset for comparisons.

```python
#Check the types of engine in our dataset
df['Engine.Type'].value_counts()
```

**Classify the engine types into four main categories based on the shared properties; 1.Reciprocating,2.Turbo for all with 'Turbo properties'3.Electric_Hybrid 4.Uknown for UNK Unknown,NONE,LR.**

```python
# Standardize text
df['Engine.Type'] = df['Engine.Type'].str.strip().str.title()

# Replace unknown values
df['Engine.Type'] = df['Engine.Type'].replace(
    ['Unknown', 'Unk', 'None', 'None ', 'Lr'],
    'Unknown'
)

# Group engine types
def engine_group(engine):
    if engine == 'Reciprocating':
        return 'Reciprocating'
    elif engine in ['Turbo Prop', 'Turbo Fan', 'Turbo Shaft', 'Turbo Jet', 'Geared Turbofan']:
        return 'Turbo'
    elif engine in ['Electric', 'Hybrid Rocket']:
        return 'Electric_Hybrid'
    else:
        return 'Unknown'

df['Engine_Group'] = df['Engine.Type'].apply(engine_group)

# Check result
df['Engine_Group'].value_counts()
```

# Clean Purpose of flight

```python
# Clean Purpose of Flight
df['Purpose_clean'] = (
    df['Purpose.of.flight']
    .str.upper()
    .str.strip()
)

# Replace missing values
df['Purpose_clean'] = df['Purpose_clean'].fillna('UNKNOWN')

df['Purpose_clean'].value_counts().head(10)
```

# Group the make column into main categories

```python
# Clean Make column
df['Make_clean'] = (
    df['Make']
    .str.upper()
    .str.strip()
)

# Count frequency
make_counts = df['Make_clean'].value_counts()

# Only group manufacturers that appear once
rare_makes = make_counts[make_counts == 1].index

# Group rare makes into OTHER
df['Make_grouped'] = df['Make_clean'].apply(
    lambda x: 'OTHER' if x in rare_makes else x
)

# Check results
df['Make_grouped'].value_counts().head(10)
```
