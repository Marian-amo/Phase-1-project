# Data Cleaning
In this part we will check fo missing values,duplicates and outliers and select the columns that are relevant to our analysis.

## Cleaning Weather Conditions
The Weather.Condition variable has labels such as 'UNK','Unk' and Nan(missing values) 
These categories are standardized into label'Unknown'

```python
#Standardize categorical values
df['Weather.Condition'] = df['Weather.Condition'].replace({
    'UNK': 'Unknown',
    'Unk': 'Unknown',
    'Nan':'Unknown'
})
df['Weather.Condition'].value_counts()

```

# Cleaning Injury Severity
In this section we will categorize the injury severity into two which are 'fatal and 'non_fatal' to simplify the column for more clarity and easy understsnding on the level of severity of injuries.

```python

# Step 1: Standardize capitalization and strip spaces
df['Injury.Severity'] = df['Injury.Severity'].astype(str).str.strip().str.capitalize()

# Step 2: Create Grouped_injury column
def clean_injury(x):
    if 'Fatal' in x:          # catches Fatal, Fatal(1), Fatal(2), etc.
        return 'Fatal'
    elif x.lower() == 'nan':   # handle NaN values converted to string
        return 'Unknown'
    else:
        return 'Non-Fatal'

df['Grouped_injury'] = df['Injury.Severity'].apply(clean_injury)

# Step 3: Check counts
print("Grouped_injury counts:")
print(df['Grouped_injury'].value_counts())
```

From the above step we can conclude that the reported number of Non fatal injuries(70063) are more as compared to reported fatal injuries(17826) in the aircraft accidents.

# Clean Aircraft specific data 

```python
#select specific columns to work with
cols = [
    'Make',
    'Model',
    'Engine.Type',
    'Purpose.of.flight',
    'Aircraft.damage',
    'Grouped_injury'
]

aircraft_df = df[cols].copy()
aircraft_df
```

```python
#clean text columns to avoid duplicates
text_cols = ['Make', 'Model', 'Engine.Type', 'Purpose.of.flight', 'Aircraft.damage']

for col in text_cols:
    aircraft_df[col] = (
        aircraft_df[col]
        .astype(str)
        .str.strip()
        .str.upper()
        .replace('NAN', 'UNKNOWN')
    )
```

```python
#group engine types
def engine_group(engine):
    if 'RECIPROCATING' in engine:
        return 'RECIPROCATING'
    elif 'TURBINE' in engine:
        return 'TURBINE'
    elif 'TURBOFAN' in engine:
        return 'TURBOFAN'
    elif 'TURBOPROP' in engine:
        return 'TURBOPROP'
    else:
        return 'OTHER / UNKNOWN'

aircraft_df['Engine_Group'] = aircraft_df['Engine.Type'].apply(engine_group)
```

```python
#Group aicraft damage according to severity
def damage_group(damage):
    if damage in ['DESTROYED']:
        return 'HIGH DAMAGE'
    elif damage in['SUBSTANTIAL']:
        return 'AVERAGE DAMAGE'
    elif damage in ['MINOR']:
        return 'LOW DAMAGE'
    else:
        return 'UNKNOWN'

aircraft_df['Damage_Group'] = aircraft_df['Aircraft.damage'].apply(damage_group)
```
